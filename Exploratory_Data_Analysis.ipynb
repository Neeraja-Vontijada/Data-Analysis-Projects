**Step 1: Import libraries and load dataset**

```python
# Importing the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#Loading the given dataset
retail= pd.read_csv("SampleSuperstore.csv")
```

```python
#Checking the dataset
retail.head()
```

```python
#Checking the no of rows and columns of the dataset
retail.shape
```

**Step 2: Dataset Analysis and Data Cleaning**

```python
# Checking summary of the dataset
retail.info()
```

```python
retail.describe()
```

```python
#Checking the missing values in dataset
retail.isnull().sum()
```

```python
# Check for missing values in the entire DataFrame
retail.isnull().sum().sum()
```

```python
# Check for any missing values (returns True or False)
retail.isnull().any().any()
```

```python
# Drop rows with any missing values
retail.dropna(inplace=True)
```

```python
# Drop rows where all columns are NaN
retail.dropna(how='all', inplace=True)
```

```python
# Drop columns with any missing values
retail.dropna(axis=1, inplace=True)
```

```python
# Drop columns where all values are missing
retail.dropna(axis=1, how='all', inplace=True)
```

```python
# First, select only numeric columns
numeric_cols = retail.select_dtypes(include='number')

# Now, compute the mean of numeric columns
mean_values = numeric_cols.mean()

# Fill NaN values with the computed means
retail[numeric_cols.columns] = retail[numeric_cols.columns].fillna(mean_values)
```

```python
retail.drop_duplicates(inplace=True)
```

**Comment: Thus,there are no missing values in the data set**


```python
#Checking the datatypes of columns
retail.dtypes
```

```python
#Checking Statistics of dataset
retail.describe()
```

```python
#Checking unique values in each columns
retail.nunique()
```

```python
# Deleting the unwanted columns
c = ['Postal Code', 'Country']   #postal code is not required for analysis
retail1 = retail.drop(columns=c,axis=1)
retail1.head()                     #Checking the dataset after removing the columns
```

**Step 3: Data Visualization**

```python
#Univariate Analysis:
#Plot histograms to understand the distribution.
#For Numerical Variables:
retail['Sales'].hist(bins=5)
```

```python
#For Categorical Variables:
#Get boxplots to detect outliers.
retail.boxplot(column='Profit')
```

```python
#For Categorical Variables:
retail['Sub-Category'].value_counts().plot(kind='bar')
```

```python
#Bivariate Analysis:
#For Two Numerical Variables:
retail.plot(kind='scatter', x='Sales', y='Profit')
```

```python
#For Numerical and Categorical Variables
#Use box plots or violin plots to compare numerical variables across categories.
sns.boxplot(x='State', y='Sales', data=retail)
```

```python
# Drawing Heatmap to identify correlation between numerical columns:
numeric_data = retail1.select_dtypes(include=['number'])
correlation=numeric_data.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation,annot=True,cmap="RdYlGn")
plt.show()
```

Inference:

1. From the heatmap, we can see that there is positive correlation of 0.48 between Sales and Profit. Thus if Sales increases, Profit increases.

2. From the heatmap, we can see that there is negative correlation of -0.22 between Profit and Discount. Thus if Discount increases, most of the times Profit goes down.

```python
# Lineplot between Profit and Discount
plt.figure(figsize=(8,6))
sns.lineplot(x='Discount',y='Profit',data=retail1)
plt.show()
```

```python
# Calculating the statewise profit
profit_sum = retail1.groupby("Sales")["Profit"].sum()
profit_sum
```

```python
# State wise Profit of United States of America
profit_sum = retail1.groupby("State")["Profit"].sum()
plt.figure(figsize=(12,6))
sns.barplot(x=profit_sum.index, y=profit_sum.values)
plt.ylabel("Profit")
plt.xticks(rotation=90)
plt.show()
```

Inference:

1. Highest profit is shown by 'California' and lowest profit is shown by 'Texas'

2. Top 4 profit giving states: 'California','New York','Washington'and'Michigan'

3. Top 4 loss giving states:'Illinois','Pennsylvania','Ohio'and'Texas'

```python
# State wise Sales of USA
sales_sum = retail1.groupby("City")["Sales"].sum()
sales_sum
```

```python
# State wise Sales of United States of America
sales_sum = retail1.groupby("City")["Sales"].sum()
plt.figure(figsize=(18,12))
sns.barplot(x=sales_sum.index, y=sales_sum.values)
plt.ylabel("Sales")
plt.xticks(rotation=90)
plt.show()
```

Inference:

1. Highest sales is shown by 'California'

2. Top 4 States with highest Sales : 'California','New York','Texas'and 'Washington'

3. Top 4 states with lowest Sales:'North Dakota','Maine','West Virginia'and'South Dakota'

Thus, Texus has high sales but low profit

```python
# State wise Avg_discount_percent of USA
Avg_discount_percent = retail1.groupby("State")["Discount"].mean()*100
Avg_discount_percent
```

```python
# State wise Avg_discount_percent of United States of America
plt.figure(figsize=(15,10))
sns.barplot(x=Avg_discount_percent.index, y=Avg_discount_percent.values)
plt.ylabel("Avg_discount_percent")
plt.xticks(rotation=90)
plt.show()
```

```python
# Distrubution of Profits of Segment across diffrent regions
plt.figure(figsize = (10,6))
sns.barplot(data = retail1 , x = "Region",y = "Profit" ,hue = "Segment")
plt.show()
```

Inference: Highest profits for each region are as follows:

South= Consumer Segment

West= Corporate Segment

East= Home Office Segment

Lowest profits for each region are as follows:

```python
# Analyzing Category wise sales, Discount, Profit:


category_data = retail1.groupby(['Category'])[['Sales', 'Discount', 'Profit']].mean()
category_data
```

```python
# Pie Charts of Category wise sales, Discount, Profit
category_data.plot.pie(subplots=True, figsize=(20, 18),autopct='%0.1f%%', labels = category_data.index)
plt.show()
```

Inference:

1)Technology Category has the maximum Sales and Profit.

2)Office Supplies category has minimum sales and Furniture category has lowest Profit.

```python
# Analyzing Ship Mode wise sales, Discount, Profit

shipmode_data = retail1.groupby(['Ship Mode'])[['Sales', 'Discount', 'Profit']].mean()
shipmode_data.plot.pie(subplots=True,figsize=(18, 20), autopct='%0.1f%%', labels = shipmode_data.index)
plt.show()
```

Inference:

Same Day Ship Mode has the highest sales.

First Class Ship mode has the highest Profit and Discount

Standard Class Ship mode has lowest sales.

Standard Class Ship mode has lowest profit.

```python
# Checking the profits of 'Sub- Category'
pivot=pd.pivot_table(retail1,index='Sub-Category',values='Profit')
pivot.plot(kind='bar')
plt.show()
```

```python
# Checking the Sales of 'Sub- Category'
pivot1=pd.pivot_table(retail1,index='Sub-Category',values='Sales')
pivot1.plot(kind='bar')
plt.show()
```
